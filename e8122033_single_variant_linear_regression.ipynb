```python
 import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv('https://raw.githubusercontent.com/Rhevathi/DATA-SCIENCE/master/ML/1_linear_reg/Exercise/canada_per_capita_income.csv')
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income')
plt.scatter(df['year'],df['per capita income (US$)'],color='red',marker='+')
```

```python
x=df.drop('per capita income (US$)',axis=1)
```

```python
x
```

```python
y=df['per capita income (US$)']
```

```python
y
```

```python
sregmod=linear_model.LinearRegression()
sregmod.fit(x,y)
```

```python
sregmod.predict([[2020]])
```

```python
sregmod.coef_
```
